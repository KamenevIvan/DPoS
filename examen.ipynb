```python
import sys

def main():
    data = sys.stdin.buffer.read().split()
    if not data:
        return
    n = int(data[0])
    prices = list(map(int, data[1:1 + n]))

    # left[i] – максимальная прибыль одной сделки, завершённой не позже дня i
    left = [0] * n
    min_price = prices[0]
    for i in range(1, n):
        left[i] = max(left[i - 1], prices[i] - min_price)
        if prices[i] < min_price:
            min_price = prices[i]

    # right[i] – максимальная прибыль одной сделки, начатой не раньше дня i
    right = [0] * n
    max_price = prices[-1]
    for i in range(n - 2, -1, -1):
        right[i] = max(right[i + 1], max_price - prices[i])
        if prices[i] > max_price:
            max_price = prices[i]

    # лучшая сумма двух непересекающихся сделок
    ans = 0
    for i in range(n):
        ans = max(ans, left[i] + right[i])

    print(ans)

if __name__ == "__main__":
    main()
```
